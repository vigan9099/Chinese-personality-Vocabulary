```python
from gensim.models import KeyedVectors
import pandas as pd
import numpy as np
import re
import emoji
data = pd.read_csv('our_embedding/腾讯.csv').drop('Unnamed: 0',axis=1)
```

```python
import gensim
import time
tic= time.time()
wvmodel_Tencent = gensim.models.KeyedVectors.load_word2vec_format('E:\词向量\Tencent_AILab_ChineseEmbedding\Tencent_AILab_ChineseEmbedding.txt', binary=False, encoding='utf-8')
toc = time.time()
print('腾讯词向量读取完毕，共耗时',toc - tic)
```

```python
# 分词个数
print(len(wvmodel_Tencent.wv.index2word ))
```

```python
sim_word = []
def similar(word,n):
    try:
        for i in wvmodel_Tencent.similar_by_word(word,n): #返回元组
            sim_word.append(i[0])
    except:
        print('没有这个词向量！')
```

```python
for k,t in enumerate(data['word']):
    print(k,'个',t)
    similar(t,20)
```

```python
def get_vec(words_list):
    my_vector_dict = {}
    for k,word in enumerate(words_list):
        print(k,'个词',word)
        word = str(word)
        try:
            my_vector_dict[word] = wvmodel_Tencent.wv.get_vector(word)
        except:
            print(word,'没有找到对应的向量')
            continue
    name_list = []
    vec_list = []
    for k in my_vector_dict:
        name_list.append(k)
        vec_list.append(my_vector_dict[k])
    return name_list,vec_list
```

```python
name_list,vec_list = get_vec(sim_word)
custom_wv = KeyedVectors(200) 
custom_wv.add(name_list, vec_list)
custom_wv.save_word2vec_format('embedding/【全】腾讯.txt')
our_embedding_by_Wiki = pd.DataFrame(vec_list)
our_embedding_by_Wiki['word']=name_list
our_embedding_by_Wiki.to_excel('our_embedding/【全】腾讯.xls')
```
